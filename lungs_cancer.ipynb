```python
##import all neccesaary Library
import numpy as np
import pandas as pd
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
```

```python
lungs_data=pd.read_csv('Dataset_csv/survey lung cancer.csv')
```

```python
lungs_data.head()
```

```python
lungs_data.shape
```

```python
lungs_data.info()
```

```python
lungs_data.describe()
```

```python
from sklearn.preprocessing import StandardScaler,LabelEncoder
```

```python
label_encod = LabelEncoder()
lungs_data['GENDER']= label_encod.fit_transform(lungs_data['GENDER']) 
lungs_data['GENDER'].unique()
```

```python
lungs_data.head()
```

```python
# 1----> Male
# 0----> Female
```

```python
lungs_data['LUNG_CANCER']= label_encod.fit_transform(lungs_data['LUNG_CANCER'])  
lungs_data['LUNG_CANCER'].unique()
```

```python
lungs_data.head()
```

```python
## 1-----> YES(LUNG_CANCER)
## 0-----> NO(LUNG_CANCER)
```

```python
X = lungs_data.drop(columns='LUNG_CANCER', axis=1)
y = lungs_data['LUNG_CANCER']
```

```python
X.info()
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
# accuracy on training data
X_train_pred = model.predict(X_train)
training_data_accu = accuracy_score(X_train_pred, y_train)
```

```python
training_data_accu
```

```python
import pickle

# Save the model
with open("lungs_cancer_model.pkl", "wb") as file:
    pickle.dump(model, file)
```
